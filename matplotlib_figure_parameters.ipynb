```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(dpi=100)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(figsize=(12,8),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(figsize=(1,8),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(figsize=(10,3),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig=plt.figure(figsize=(10,3),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_figure.png')
```

```python
fig=plt.figure(figsize=(2,2),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_figure.png')
```

```python
fig=plt.figure(figsize=(2,2),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_figure.png',bbox_inches='tight')
```
